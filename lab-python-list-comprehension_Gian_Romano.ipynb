````python
products = ['t-shirt', 'mug', 'hat', 'book', 'keychain']

# Function to load quntities in the inventory
def initialize_inventory(products):
    inventory = {}
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) 
                 for product in products}
    return inventory

# Function to record customer order
def get_customer_orders():
    product_to_buy = int(input("Enter the number of customer orders: "))
    customer_orders = [input("Enter the name of a product that a customer wants to order: ") 
                       for product in range(product_to_buy)]
    return customer_orders

# Function to Calculate Total Price
def calculate_total_price(customer_orders):
    price_of_purchase = [float(input(f"Enter the price of {product}: ")) 
                         for product in customer_orders]
    total = sum(price_of_purchase)   
    return total 

# Function to update inventory
def update_inventory(customer_orders,inventory):
    for sold_product in customer_orders:
        if sold_product in inventory:
         inventory[sold_product]  =  (inventory[sold_product] - 1)
    
    inventory = {product: qty for product, qty in inventory.items() if qty > 0}  
    return inventory

# Function to calculate order statistics
def calculate_order_statistics(customer_orders,products):
    total_products_ordered = len(customer_orders)
    total_products = len(products)
    percentage_of_unique_products_ordered = (total_products_ordered/total_products)*100
    order_statistics = (total_products_ordered,percentage_of_unique_products_ordered)
    return order_statistics

# Function to print order statistics   
def print_order_statistics(order_statistics):
    print("Order Statistics:")
    print(f"Total Products Ordered: {order_statistics[0]}")
    print(f"Percentage of Products Ordered: {order_statistics[1]}%")
    
# Function to print updated inventory
def print_updated_inventory(inventory):
    print("Updated Inventory: ")
    [print(f"{product} : {quantity}") for product, quantity in inventory.items()] # Assited by ChatGpt

# Function to show total for purchase
def total_purchase(total):
    print(f"Total Price: {total}")

# Main function 
def main():
    inventory = initialize_inventory(products)
    customer_orders = get_customer_orders()  
    order_statistics = calculate_order_statistics(customer_orders,products)
    print_order_statistics(order_statistics)
    update_inventory(customer_orders,inventory)
    print_updated_inventory(inventory)
    total = calculate_total_price(customer_orders)
    total_purchase(total)
    
main()
````
